Step 3 - Seperate X and Y(loan_status)

```python
X = df.drop(columns = ["id", "loan_status"])
Y = df["loan_status"]
```

```python
X.head()
```

```python
Y.head()
```

Step 4 - Apply train test split

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```


Step 5 - Preprocessing

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
)
```

```python
num_cols = X.select_dtypes(include="number").columns.tolist()
cat_cols = X.select_dtypes(include="object").columns.tolist()

pre = ColumnTransformer(
    [
        ("num", num_pipe, num_cols),
        ("cat", cat_pipe, cat_cols)
    ]
).set_output(transform="pandas")

pre.fit(xtrain)
```

```python
xtrain_pre = pre.transform(xtrain)
xtrain_pre.head()
```

```python
xtest_pre = pre.transform(xtest)
xtest_pre.head()
```

# Bagging model - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(
    n_estimators=1,
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="gini"
)

model1.fit(xtrain_pre, ytrain)
```

```python
model1.score(xtrain_pre, ytrain)
```

```python
model1.score(xtest_pre, ytest)
```

# Hyperparameter tune he random forest model

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    "max_depth": [2, 3, 4, 5],
    "n_estimators": [10, 50, 100, 200]
}

rf = RandomForestClassifier(random_state=42)

rscv_rf = RandomizedSearchCV(rf, params, cv=5, scoring="f1_macro")

rscv_rf.fit(xtrain_pre, ytrain)
```

```python
rscv_rf.best_params_
```

```python
rscv_rf.best_score_
```

```python
best_rf = rscv_rf.best_estimator_
best_rf
```

# Evaluate the model

```python
best_rf.score(xtrain_pre, ytrain)
```

```python
best_rf.score(xtest_pre, ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_rf, xtest_pre, ytest)
```

```python
from sklearn.metrics import classification_report

ypred_test = best_rf.predict(xtest_pre)
print(classification_report(ytest, ypred_test))
```

```python
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(best_rf, xtest_pre, ytest)
```
